```python
import torch

print(f"PyTorch version: {torch.__version__}")
print(f"Is CUDA (GPU) available? {torch.cuda.is_available()}")
print(f"Is MPS (Mac GPU) available? {torch.backends.mps.is_available()}")

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_id = "allenai/OLMo-2-0425-1B-Instruct" # Or the base model
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,  # Directed: Use bfloat16
    device_map="auto"            # Automatically handles GPU/CPU split
)
```

```python
prompt = "Explain why the sky is blue."
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(**inputs, max_new_tokens=100)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
# 3. Prepare the Input
prompt = "Explain the concept of quantum entanglement in one sentence."
inputs = tokenizer(prompt, return_tensors="pt").to(device)
inputs = inputs.to(model.device)

# 4. Generate the response
with torch.no_grad():
    output_tokens = model.generate(
        **inputs, 
        max_new_tokens=50,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )

# 5. Decode and Print
response = tokenizer.decode(output_tokens[0], skip_special_tokens=True)
print(f"\nModel Response:\n{response}")
```
